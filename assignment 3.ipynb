```python
import random

def load_word(file):
    with open(file, 'r') as file:
        words = file.read().splitlines()
    return words

def sort_word(words):
    
    sorted_words = {}
    for word in words:
        length = len(word)
        sorted_words.setdefault(length, []).append(word)
    return sorted_words

import random

def display_hangman(wrong_guess_time):
    hangman=[
        '''
        --------
        |    
        |
        |
        |
        |
        |
        ''',
        '''
        ---------
        |     |
        |
        |
        |
        |
        |
        ''',
        '''
        ---------
        |     |_
        |     |_|
        |     |
        |
        |
        |
        
        
        ''',
         '''
        ---------
        |     |_
        |     |_|
        |   / | 
        |
        |
        |
        ''',
         '''
        ---------
        |     |_
        |     |_|
        |   / |\\
        |
        |
        |
        ''',
        '''
        ---------
        |     |_
        |     |_|
        |   / |\\
        |     |
        |   /
        |
        ''',
        '''
        ---------
        |     |_
        |     |_|
        |   / |\\
        |     |
        |   /   \\
        |
        ''',
        
        
        
        
        
    ]
    print (hangman[wrong_guess_time])

def play_the_game(word):
    
    guessed_letters = [ ]
    wrong_guess_time = 0
    max_incorrect_guesses = 6
    score = 0
    Current=[]

    print('Start!')
    
    while wrong_guess_time < max_incorrect_guesses:
        print(display_hangman(wrong_guess_time))
        print("Current word:", '_'*len(word))
        print("Guessed letters:", guessed_letters)
        guess = input("Guess a letter: ")

       
        
        guessed_letters.append(guess)

        if guess in word:
            print(f"Yes'{guess}' is in the word.")
            score += 10
            Current.append(guess)
            print('Current word',Current,'_'*(len(word)-len(Current)))
        else:
            print(f"Sorry, '{guess}' is not in the word.")
            wrong_guess_time += 1
            score -= 5

    if len(Current)==len(word):
        print('Yes!You succeed!The word is::',word,'Your score:',score)
    else:
        print(display_hangman(wrong_guess_time))
        print(f"Sorry, you ran out of guesses. The word was: {word}. Your score: {score} points.")

def main():
    words = load_word('words.txt')
    sorted_words = sort_word(words)

    while True:
        length_choice = int(input("Choose the word length (6 or more): "))
        word = random.choice(sorted_words[length_choice])
        play_the_game(word)
        
        
        play_again = input("Do you want to play again? (y/n): ")
        if play_again != 'y':
            break

if __name__ == "__main__":
    main()
```
